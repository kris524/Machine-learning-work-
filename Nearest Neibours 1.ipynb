## 1 nearest neiboughr with iris data 


```python
import math
import numpy as np
from sklearn.datasets import load_iris
from sklearn.metrics import zero_one_loss
iris = load_iris()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'],iris['target'],random_state=211)



labels = []
for row_i in X_test:
    empty_list=[]
    for row_j in X_train:
        d_2 = (row_i - row_j)**2
        d = np.sqrt(sum(d_2))#here we get a single number distance from X_test0 to X_trainj

   
        empty_list.append(d)
  
        array_d = np.array(empty_list)   
        
#-----------------------------------------------------------------------------------------------

        
    #find the min distance to X_test[j]    
    min_value = math.inf   
    for n in range(array_d.size):
        if min_value > array_d[n]:
        
            min_value = array_d[n]
            #print('index of mimium',n )# we get the index 71 for row_i = 0.
            label = y_train[n] #take the label from 71 and assign it to the y_test
               
   # print('The minimum value from the array is:',min_value)# we get the distance smallest to X_test0

   # print('The label for X_test[{}] is {}'.format(row_i,label))
   # print('---------------------------------------------')
    #print('The arary which is closest to X_test[0] is: ',min_value)
#-----------------------------------------------------------------------------------------------
#This part I will show the error rate 

    labels.append(label)
#print('The list of labels is here:',labels)
y_pred_array = np.array(labels)
error_rate = zero_one_loss(y_test, y_pred_array)
#print('The test error rate is', error_rate)
```
